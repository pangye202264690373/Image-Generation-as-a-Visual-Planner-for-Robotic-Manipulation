```python
import json

jsonl_path = r"F:\PangYe\text_in\parsed_instructions.jsonl"
data = []
bad_lines = 0

with open(jsonl_path, "r", encoding="utf-8") as f:
    for i, line in enumerate(f, 1):
        line = line.strip()
        if not line:
            continue  # 跳过空行
        try:
            data.append(json.loads(line))
        except json.JSONDecodeError as e:
            bad_lines += 1
            print(f"[WARN] 第 {i} 行 JSON 解析失败: {e}")
            print("  内容片段:", line[:200], "...\n")

print(f"✅ 成功读取 {len(data)} 条记录；跳过坏行 {bad_lines} 条。")

if data:
    print("示例记录：")
    print(json.dumps(data[0], indent=2, ensure_ascii=False))
```
